```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
y1=np.cos(x)
plt.plot(x,y1,color='blue')
plt.scatter(x,y,color='red')
plt.bar(x,y,color='red')
plt.xlabel('samples')
plt.xlabel('amplitude')
plt.title('sin')
plt.legend(['sine','cos'])
```
